```python
def clean_data(df):
    # Drop rows with missing data in columns: 'country', 'agent'
    df = df.dropna(subset=['country', 'agent'])
    # Drop column because of too many missing value: 'company'
    df = df.drop(columns=['company'])
    # Drop duplicates and reset index
    df.drop_duplicates(inplace=True)
    df.reset_index(drop=True, inplace=True)
    return df

df_clean = clean_data(df.copy())
df_clean.head()
```

```python
# Set style for better visualizations
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = [12, 6]
```

```python
# Create total_stay feature
df_clean['total_stay'] = df_clean['stays_in_weekend_nights'] + df_clean['stays_in_week_nights']
```

```python
def plot_numeric_distributions(df):
    numeric_cols = ['lead_time', 'booking_changes', 'total_stay',
                   'previous_cancellations', 'previous_bookings_not_canceled', 'adr']

    plt.figure(figsize=(15, 10))

    for i, col in enumerate(numeric_cols, 1):
        plt.subplot(2, 3, i)

        # Adjust binning based on data characteristics
        if col in ['booking_changes', 'previous_cancellations', 'previous_bookings_not_canceled']:
            # For discrete variables with small ranges
            bins = int(df[col].max() - df[col].min() + 1)
            sns.histplot(df[col], kde=False, bins=bins)
        elif col == 'adr':
            # For ADR, limit x-axis to 99th percentile for better visibility
            # but still include all data in the histogram
            sns.histplot(df[col], kde=True, bins=50)
            plt.xlim(0, df[col].quantile(0.99))
        elif col == 'lead_time':
            # For lead_time, use more bins and limit x-axis to 95th percentile
            sns.histplot(df[col], kde=True, bins=50)
            plt.xlim(0, df[col].quantile(0.95))
        else:
            # For other variables
            sns.histplot(df[col], kde=True, bins=30)

        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel('Count')

        # Add mean and median lines
        mean_val = df[col].mean()
        median_val = df[col].median()
        plt.axvline(mean_val, color='red', linestyle='--', alpha=0.5, label=f'Mean: {mean_val:.2f}')
        plt.axvline(median_val, color='green', linestyle='--', alpha=0.5, label=f'Median: {median_val:.2f}')

        # Add legend
        plt.legend(fontsize='small')

    plt.tight_layout()
    plt.show()
```

```python
# 2. Boxplots for Numerical Variables (to detect outliers)
def plot_boxplots(df):
    numeric_cols = ['lead_time', 'booking_changes', 'total_stay',
                   'previous_cancellations', 'previous_bookings_not_canceled', 'adr']

    plt.figure(figsize=(15, 5))
    sns.boxplot(data=df[numeric_cols])
    plt.xticks(rotation=45)
    plt.title('Boxplots of Numerical Variables')
    plt.show()
```

```python
# 3. Categorical Variables Distribution
def plot_categorical_distributions(df):
    categorical_cols = ['hotel', 'arrival_date_month', 'market_segment', 'deposit_type']

    plt.figure(figsize=(15, 10))
    for i, col in enumerate(categorical_cols, 1):
        plt.subplot(2, 2, i)
        sns.countplot(data=df, x=col)
        plt.xticks(rotation=45)
        plt.title(f'Distribution of {col}')
    plt.tight_layout()
    plt.show()
```

```python
# 4. Correlation Heatmap
def plot_correlation_heatmap(df):
    numeric_cols = ['lead_time', 'booking_changes', 'total_stay',
                   'previous_cancellations', 'previous_bookings_not_canceled', 'adr']

    plt.figure(figsize=(10, 8))
    sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', center=0)
    plt.title('Correlation Heatmap of Numerical Variables')
    plt.show()
```

```python
# 5. Cancellation Analysis by Categories
def plot_cancellation_analysis(df):
    categorical_cols = ['hotel', 'arrival_date_month', 'market_segment', 'deposit_type']

    plt.figure(figsize=(15, 10))
    for i, col in enumerate(categorical_cols, 1):
        plt.subplot(2, 2, i)
        cancellation_rates = df.groupby(col)['is_canceled'].mean() * 100
        cancellation_rates.plot(kind='bar')
        plt.title(f'Cancellation Rate by {col}')
        plt.ylabel('Cancellation Rate (%)')
        plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# 6. Average Daily Rate Analysis
def plot_adr_analysis(df):
    plt.figure(figsize=(15, 5))

    # ADR by Hotel Type
    plt.subplot(1, 3, 1)
    sns.boxplot(data=df, x='hotel', y='adr')
    plt.title('ADR by Hotel Type')

    # ADR by Market Segment
    plt.subplot(1, 3, 2)
    sns.boxplot(data=df, x='market_segment', y='adr')
    plt.xticks(rotation=45)
    plt.title('ADR by Market Segment')

    # ADR by Deposit Type
    plt.subplot(1, 3, 3)
    sns.boxplot(data=df, x='deposit_type', y='adr')
    plt.title('ADR by Deposit Type')

    plt.tight_layout()
    plt.show()
```

```python
# 7. Lead Time Analysis
def plot_lead_time_analysis(df):
    plt.figure(figsize=(15, 5))

    # Lead Time Distribution by Hotel
    plt.subplot(1, 3, 1)
    sns.boxplot(data=df, x='hotel', y='lead_time')
    plt.title('Lead Time by Hotel Type')

    # Lead Time Distribution by Market Segment
    plt.subplot(1, 3, 2)
    sns.boxplot(data=df, x='market_segment', y='lead_time')
    plt.xticks(rotation=45)
    plt.title('Lead Time by Market Segment')

    # Lead Time vs Cancellation
    plt.subplot(1, 3, 3)
    sns.boxplot(data=df, x='is_canceled', y='lead_time')
    plt.title('Lead Time by Cancellation Status')

    plt.tight_layout()
    plt.show()
```

```python
# Execute all visualizations
print("1. Distribution of Numerical Variables")
plot_numeric_distributions(df_clean)

print("\n2. Boxplots for Numerical Variables")
plot_boxplots(df_clean)

print("\n3. Categorical Variables Distribution")
plot_categorical_distributions(df_clean)

print("\n4. Correlation Heatmap")
plot_correlation_heatmap(df_clean)

print("\n5. Cancellation Analysis by Categories")
plot_cancellation_analysis(df_clean)

print("\n6. Average Daily Rate Analysis")
plot_adr_analysis(df_clean)

print("\n7. Lead Time Analysis")
plot_lead_time_analysis(df_clean)
```

```python
# 1. Load and preprocess data
def preprocess_data(df, sample_size=3000, random_state=42):
    # Random sampling
    df_sampled = df.sample(n=sample_size, random_state=random_state)

    # Handle ADR outliers
    Q1 = df_sampled['adr'].quantile(0.25)
    Q3 = df_sampled['adr'].quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + 3 * IQR
    df_sampled.loc[df_sampled['adr'] > upper_bound, 'adr'] = upper_bound

    # Transform lead_time
    df_sampled['lead_time_log'] = np.log1p(df_sampled['lead_time'])

    return df_sampled

# Process
df_processed = preprocess_data(df_clean, sample_size=3000)

# Print some info to verify sampling
print("Original dataset size:", len(df_clean))
print("Sampled dataset size:", len(df_processed))

# Verify distributions are similar
print("\nCancellation Rate Comparison:")
print("Original:", df_clean['is_canceled'].mean())
print("Sampled:", df_processed['is_canceled'].mean())
```

```python
# Define features
numeric_features = [
    'lead_time_log',
    'booking_changes',
    'total_stay',
    'previous_cancellations',
    'previous_bookings_not_canceled',
    'adr'
]

categorical_features = [
    'hotel',
    'arrival_date_month',
    'market_segment',
    'deposit_type'
]
```

```python
# 2. Handle categorical features with weighted encoding for market_segment
market_segment_weights = {
    'Online TA': 0.35,
    'Groups': 0.31,
    'Direct': 0.18,
    'Corporate': 0.18,
    'Complementary': 0.12,
    'Aviation': 0.20,
    'Offline TA/TO': 0.15
}

le = LabelEncoder()
df_encoded = df_processed.copy()

# Special handling for market_segment
df_encoded['market_segment_weight'] = df_encoded['market_segment'].map(market_segment_weights)

# Encode other categorical features
for cat_feature in [f for f in categorical_features if f != 'market_segment']:
    df_encoded[cat_feature] = le.fit_transform(df_encoded[cat_feature])
```

```python
# 3. Combine features
features_for_clustering = numeric_features + [f for f in categorical_features if f != 'market_segment'] + ['market_segment_weight']
X = df_encoded[features_for_clustering]
```

```python
# 4. Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=features_for_clustering)
```

```python
#################### K-MEANS ####################
# Before Feature Selection
def kmeans_clustering_before(X_scaled):
    """Model for K-means before feature selection"""
    model_kmeans = KMeans(random_state=42)
    visualizer = KElbowVisualizer(model_kmeans, k=(2,10))
    visualizer.fit(X_scaled)
    optimal_k = visualizer.elbow_value_
    if optimal_k is None:
        optimal_k = 3
    kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_scaled)
    return labels, optimal_k

def evaluate_kmeans_before(X_scaled, labels):
    """Evaluate K-means before feature selection"""
    silhouette_avg = silhouette_score(X_scaled, labels)
    print(f"K-means before feature selection - Silhouette Score: {silhouette_avg:.3f}")
    return silhouette_avg

def visualize_kmeans_before(X_pca, labels, k, silhouette_score):
    """Visualize K-means before feature selection"""
    plt.figure(figsize=(10, 6))
    scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis')
    plt.title(f'K-means Clustering Before Feature Selection\nk={k}, Silhouette Score: {silhouette_score:.3f}')
    plt.xlabel('First Principal Component')
    plt.ylabel('Second Principal Component')
    plt.colorbar(scatter)
    plt.show()

# After Feature Selection
def kmeans_clustering_after(X_selected):
    """Model for K-means after feature selection"""
    model_kmeans = KMeans(random_state=42)
    visualizer = KElbowVisualizer(model_kmeans, k=(2,10))
    visualizer.fit(X_selected)
    optimal_k = visualizer.elbow_value_
    if optimal_k is None:
        optimal_k = 3
    kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_selected)
    return labels, optimal_k

def evaluate_kmeans_after(X_selected, labels):
    """Evaluate K-means after feature selection"""
    silhouette_avg = silhouette_score(X_selected, labels)
    print(f"K-means after feature selection - Silhouette Score: {silhouette_avg:.3f}")
    return silhouette_avg

def visualize_kmeans_after(X_pca_selected, labels, k, silhouette_score):
    """Visualize K-means after feature selection"""
    plt.figure(figsize=(10, 6))
    scatter = plt.scatter(X_pca_selected[:, 0], X_pca_selected[:, 1], c=labels, cmap='viridis')
    plt.title(f'K-means Clustering After Feature Selection\nk={k}, Silhouette Score: {silhouette_score:.3f}')
    plt.xlabel('First Principal Component')
    plt.ylabel('Second Principal Component')
    plt.colorbar(scatter)
    plt.show()
```

### **a.2 DBSCAN Clustering**

```python
#################### DBSCAN ####################
# Before Feature Selection
def dbscan_clustering_before(X_scaled):
    """Model for DBSCAN before feature selection"""
    neighbors = NearestNeighbors(n_neighbors=2)
    neighbors_fit = neighbors.fit(X_scaled)
    distances, _ = neighbors_fit.kneighbors(X_scaled)
    distances = np.sort(distances[:, 1])
    eps_value = np.percentile(distances, 95)
    dbscan = DBSCAN(eps=eps_value, min_samples=5)
    labels = dbscan.fit_predict(X_scaled)
    return labels, eps_value

def evaluate_dbscan_before(X_scaled, labels):
    """Evaluate DBSCAN before feature selection"""
    if len(set(labels)) > 1:
        silhouette_avg = silhouette_score(X_scaled, labels)
        print(f"DBSCAN before feature selection - Silhouette Score: {silhouette_avg:.3f}")
        return silhouette_avg
    return 0

def visualize_dbscan_before(X_pca, labels, eps, silhouette_score):
    """Visualize DBSCAN before feature selection"""
    plt.figure(figsize=(10, 6))
    scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis')
    plt.title(f'DBSCAN Clustering Before Feature Selection\neps={eps:.3f}, Silhouette Score: {silhouette_score:.3f}')
    plt.xlabel('First Principal Component')
    plt.ylabel('Second Principal Component')
    plt.colorbar(scatter)
    plt.show()

# After Feature Selection
def dbscan_clustering_after(X_selected):
    """Model for DBSCAN after feature selection"""
    neighbors = NearestNeighbors(n_neighbors=2)
    neighbors_fit = neighbors.fit(X_selected)
    distances, _ = neighbors_fit.kneighbors(X_selected)
    distances = np.sort(distances[:, 1])
    eps_value = np.percentile(distances, 95)
    dbscan = DBSCAN(eps=eps_value, min_samples=5)
    labels = dbscan.fit_predict(X_selected)
    return labels, eps_value

def evaluate_dbscan_after(X_selected, labels):
    """Evaluate DBSCAN after feature selection"""
    if len(set(labels)) > 1:
        silhouette_avg = silhouette_score(X_selected, labels)
        print(f"DBSCAN after feature selection - Silhouette Score: {silhouette_avg:.3f}")
        return silhouette_avg
    return 0

def visualize_dbscan_after(X_pca_selected, labels, eps, silhouette_score):
    """Visualize DBSCAN after feature selection"""
    plt.figure(figsize=(10, 6))
    scatter = plt.scatter(X_pca_selected[:, 0], X_pca_selected[:, 1], c=labels, cmap='viridis')
    plt.title(f'DBSCAN Clustering After Feature Selection\neps={eps:.3f}, Silhouette Score: {silhouette_score:.3f}')
    plt.xlabel('First Principal Component')
    plt.ylabel('Second Principal Component')
    plt.colorbar(scatter)
    plt.show()
```
